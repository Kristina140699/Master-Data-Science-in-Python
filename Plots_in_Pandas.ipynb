# Python Matplotlib
### Matplotlib is a python library used for data visualization. You can create bar-plots, histograms and a lot more with matplotlib.

**Key Features of Matplotlib**:
1. **Versatility**: Matplotlib can generate a wide range of plots, including line plots, scatter plots, bar plots, histograms, pie charts, and more.
2. **Customization**: It offers extensive customization options to control every aspect of the plot, such as line styles, colors, markers, labels, and annotations.
3. **Integration with NumPy**: Matplotlib integrates seamlessly with NumPy, making it easy to plot data arrays directly.
4. **Publication Quality**: Matplotlib produces high-quality plots suitable for publication with fine-grained control over the plot aesthetics.
5. **Extensible**: Matplotlib is highly extensible, with a large ecosystem of add-on toolkits and extensions like Seaborn, Pandas plotting functions, and Basemap for geographical plotting.
6. **Cross-Platform**: It is platform-independent and can run on various operating systems, including Windows, macOS, and Linux.
7. **Interactive Plots**: Matplotlib supports interactive plotting through the use of widgets and event handling, enabling users to explore data dynamically.

```python
import numpy as np
from matplotlib import pyplot as plt
```

## Line Plot

```python
x = np.arange(1,21)
x
```

```python
y= 2*x
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
# Adding Title and labels
plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.show()
```

```python
# Changing the Line Aesthetics
plt.plot(x,y, color='green', linestyle=":", linewidth=2)
plt.show()
```

```python
plt.plot(x,y, color='black', linestyle="-.", linewidth=4)
plt.title("Asthetics")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
# Adding two lines in the same plot
x = np.arange(1,26)
y1 = 2*x
y2 = 3*x
```

```python
plt.plot(x, y1, color='g', linestyle=":", linewidth=2)
plt.plot(x, y2, color='blue', linestyle="-.", linewidth=3)
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

```python
# Adding sub-plots
x = np.arange(1,11)
y1 = 2*x
y2 = 3*x

# subplot(vertical_no, horizontal_no,index/position)
plt.subplot(1,2,1)
plt.plot(x,y1, color='black', linestyle=":", linewidth=2)

plt.subplot(1,2,2)
plt.plot(x, y2, color='r', linestyle='--', linewidth=2)

plt.show()
```

```python
# Adding sub-plots
x = np.arange(1,11)
y1 = 2*x
y2 = 3*x

# subplot(vertical_no, horizontal_no,index/position)
plt.subplot(2,1,1)
plt.plot(x,y1, color='black', linestyle=":", linewidth=2)

plt.subplot(2,1,2)
plt.plot(x, y2, color='r', linestyle='--', linewidth=2)

plt.show()
```

## Bar Plot

```python
student = {"Bob": 87, "Matt":56, "Sam":97}
names = list(student.keys())
values = list(student.values())

plt.bar(names, values)
plt.show()
```

```python
students = {'Erik':69, 'Meera':90, 'Arina': 78, 'Patrik': 94}
names = list(students.keys())
values = list(students.values())

plt.bar(names, values)
plt.title('Bar Plot')
plt.xlabel('Names')
plt.ylabel('Marks')
plt.show()
```

```python
plt.bar(names, values, color='purple')
plt.title('Bar Plot')
plt.xlabel('Names')
plt.ylabel('Marks')
plt.grid(True)
plt.show()
```

```python
plt.barh(names, values, color='yellow')
plt.title('Bar Plot')
plt.xlabel('Marks')
plt.ylabel('Names')
plt.grid(False)
plt.show()
```

## Scatter Plot

```python
# Creating a basic scatter-plot
x = [10, 20, 30, 40, 50, 60, 70, 80, 90]
y = [8, 1, 7, 2, 0, 3, 7, 3, 2]

plt.scatter(x,y)
plt.show()
```

```python
# Changing the Mark Aesthetics
x = [10, 20, 30, 40, 50, 60, 70, 80, 90]
y = [7, 19, 12, 6, 9, 17, 20, 13, 19]

plt.scatter(x,y, marker='*', c='red', s=100)
plt.show()
```

```python
x = [10, 20, 30, 40, 50, 60, 70, 80, 90]
a = [8, 1, 7, 2, 0, 3, 7, 3, 2]
b = [7, 19, 12, 6, 9, 17, 20, 13, 19]
```

```python
# Two markers
# Adding two markers in the same plot

plt.scatter(x,a, c='blue', marker = '*', s=100)
plt.scatter(x,b, c='black', marker = 'o', s=20)
plt.show()
```

```python
# Adding sub-plots

plt.subplot(1,2,1) #divinding the sapce vertically into 2 and this is the first plot
plt.scatter(x,a, marker='*', c='purple', s=100)

plt.subplot(1,2,2) #this is the second plot
plt.scatter(x,b, marker='.', c='b', s=100)

plt.show()
```

```python
# Adding sub-plots

plt.subplot(2,1,1) #divinding the sapce vertically into 2 and this is the first plot
plt.scatter(x,a, marker='*', c='black', s=100)

plt.subplot(2,1,2) #this is the second plot
plt.scatter(x,b, marker='.', c='purple', s=100)

plt.show()
```
